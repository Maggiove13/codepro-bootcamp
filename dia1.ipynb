## Variables:

Las variables son espacios en la memoria de la computadora en los cuales podemos guardar datos mientras se ejecuta un programa.
Cuando el programa se est√° ejecutando, la computadora necesita recordar informaci√≥n: n√∫meros, textos, resultados de operaciones, etc. Para eso se utilizan las variables.
A estos espacios de memoria les asignamos un nombre, y mediante ese nombre podemos acceder m√°s tarde al valor que contienen y utilizarlo en distintas partes del programa.

####  Reglas para nombrar variables

Algunas reglas y buenas pr√°cticas para nombrar variables en Python:

- ‚úÖ Puede incluir letras, n√∫meros y guiones bajos (_).
- ‚úÖ Debe empezar con una letra o gui√≥n bajo, nunca con un n√∫mero.
- ‚úÖ Evitar usar palabras reservadas de Python (como print, def, if, etc).
- ‚úÖ Usar nombres descriptivos (no pongas x, a, b si no hace falta!).
En programacion un espacio vacio es un caracter, por eso al escribir variables no deben existir espacios vacios


```python
# Orden de ejecucion:

fruta = "naranja"
print(fruta)

cantidad_de_frutas = 4
print(cantidad_de_frutas)

fruta = "banana"
print(fruta)
```

## Operaciones aritm√©ticas / con variables

Tambi√©n podemos realizar operaciones matem√°ticas, como en las calculadoras.
Python lee los c√≥digos de izquierda a derecha, y sigue reglas de precedencia, como en matem√°ticas

### Operadores aritm√©ticos
- `+` - Suma
- `-` - Resta
- `*` - Multiplicaci√≥n
- `/` - Divisi√≥n
- `//` - Divisi√≥n entera (sin decimales)
- `%` - Resto (m√≥dulo)  

```python
producto1 = 'papas fritas'
costo_producto1 = 15000
producto2 = 'coca cola'
costo_producto2 = 14000
capital = 50000
dolar = 6700

# Realizamos la suma
costo_total = costo_producto1 + costo_producto2
print(costo_total)

# REalizamos la resta
vuelto = capital - costo_total
print(vuelto)

# Realizamos la division
vuelto_dolar = vuelto//dolar
print(vuelto_dolar)
```

### Que es un algoritmo

Un algoritmo es una serie de pasos o instrucciones definidos y ordenados para resolver un problema o completar una tarea espec√≠fica, muy de la mano de lo que es el pensamiento computacional. Los algoritmos se utilizan para describir c√≥mo se deben llevar a cabo procesos de manera precisa y detallada, debe tener un n√∫mero finito de pasos y terminar en alg√∫n momento.


## Funciones 

Las funciones son un conjunto de instrucciones de una operacion que quisieramos volver a repetir en diferentes partes de nuestro programa. 

Las funciones existen para permitirnos re utilizar codigo que ya escribimos anteriormente, sin la necesidad de volver a escribirlo. 

Imagin√° que est√°s preparando un s√°ndwich. Siempre segu√≠s los mismos pasos, entonces en vez de escribir esas instrucciones cada vez, pod√©s agruparlas en una funci√≥n y simplemente "llamarla" cuando lo necesites.



```python
# Aca definimos la funcion
def preparar_sandwich():
    print("Preparar sandwich")
    print("Ya el sandwich esta casi listo")
# Aqui termina la funcion

# Aca le Llamamos a la funcion
preparar_sandwich() # Primera llamada
preparar_sandwich() # segunda vez
preparar_sandwich() # tercera vez
```

### Funciones con parametros y argumentos:

A veces queremos ejecutar "casi" la misma funci√≥n varias veces. Por ejemplo, si somos vendedores de autos...

Las funciones pueden recibir informaci√≥n desde afuera, para poder hacer cosas diferentes cada vez que se usan. Esa informaci√≥n se pasa a trav√©s de par√°metros (cuando definimos la funci√≥n) y argumentos (cuando llamamos a la funci√≥n).


```python
# # Definimos una funcion con el parametro "ingrediente"
def preparar_sandwich(ingrediente):
    print(f"Preparar sandwich de {ingrediente}")


# # Ejecurtar la funcion con el argumento
preparar_sandwich("pollo")
preparar_sandwich("milanesa")
preparar_sandwich("atun")

ingrediente1 = "milanesa"
ingrediente2 = "pollo"

# Definimos la funcion
def preparar_sandwich(ingrediente1, ingrediente2):
    print(f"Preparar sandwich de {ingrediente1} con {ingrediente2}")

# Llamar a la funcion con los argumentos
preparar_sandwich(ingrediente1, ingrediente2)
```

## Challenge ü§∫ - 25m

Escribir una funcion que nos pregunte las caracteristicas de un elemento a comprar e imprima el elemento con sus caracteristicas. Debes usar parametros y argumentos

```python
# Solucion de Sebas:
# Escribir una funcion que nos pregunte las caracteristicas de un elemento a comprar e imprima el elemento con sus caracteristicas.

# DEFINIR  UNA FUNCION: NO TEINE VALOR
def mostrar_producto(elemento, primera_caracteristica, segunda_caracteristica, tercera_caracteristica):
    print(f"El elemento a comprar es {elemento}, y cuenta con las siguientes caracteristicas: {primera_caracteristica}, {segunda_caracteristica}, {tercera_caracteristica}")

def producto_a_comprar():
    elemento = input("Ingrese el elemento a comprar: ")
    primera_caracteristica = input("Ingrese una caracteristica del elemento a comprar: ")
    segunda_caracteristica = input("Ingrese otra caracteristica del elemento a comprar: ")
    tercera_caracteristica = input("Ingrese una ultima caracteristica del elemento a comprar: ")

    mostrar_producto(elemento, primera_caracteristica, segunda_caracteristica, tercera_caracteristica)

producto_a_comprar()
```
