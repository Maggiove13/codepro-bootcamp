## Diccionarios
Un diccionario es una colecci√≥n de pares formados por una clave y un valor asociado a la clave. Se construyen poniendo los pares entre llaves `{ }` separados por comas, y separando la clave del valor con dos puntos :.

Se caracterizan por:

- No tienen orden.
- Pueden contener elementos de distintos tipos.
- Son mutables, es decir, pueden alterarse durante la ejecuci√≥n de un programa.
- Las claves son √∫nicas, es decir, no pueden repetirse en un mismo diccionario
- Los valores de un diccionario pueden ser de cualquier tipo de dato: int, float, bool, list, dict, etc.
- Son iterables

```python
# Crear un diccionario
mi_contacto = {"nombre": "Fulana",
            "apellido": "Britez",
            "email": "alguin@homail.com",
            "telefono": "0980111111"}
print(mi_contacto)

# Acceder a una clave del diccionario
mi_contacto['email']
print(mi_contacto['email'])

producto = {
    "nombre": "bateria",
    "precio": 450000, #integer
    "stock": True # bool
    
}
# Acceder a una clave del diccionario
print(producto["precio"])

# Modificar un elemento del diccionario
mi_contacto["email"] = 'penguin@homail.com'
print(mi_contacto)

# Usar metodos del diccionario.
print(mi_contacto.keys()) # Devuelve todas las claves del diccionario
print(mi_contacto.values()) # Devuelve todas las claves del diccionario
print(mi_contacto.items()) # Devuelve el par

# Anhadir mas datos al diccionario
mi_contacto['ciudad'] = 'Asuncion'
print(mi_contacto)

mi_contacto["nombre"] = 'Jose'
print(mi_contacto)

# Eliminar el ultimo elemento (clave-valor) del diccionario
mi_contacto.popitem()
print(mi_contacto)

# Eliminar datos del diccionario por clave:
del mi_contacto["apellido"]
print(f"Este es el diccionario con la clave eliminada {mi_contacto}")

# Los diccionarios son iuterables
for valores in mi_contacto.values():
    print(valores)

for claves in mi_contacto.keys():
    print(claves)

for clave, valor in mi_contacto.items():
    print(clave, valor)

```

## Programaci√≥n orientada a objetos

Python tambi√©n nos permite representar cosas del mundo real, m√°s fielmente por medio de estructuras de datos llamadas **objetos**.

Tanto sus atributos (caracter√≠sticas y propiedades) como sus m√©todos (funciones que nos permiten interactuar con el objeto) se definen en una `clase`.

Esta `clase` es como una plantilla para los objetos que queremos crear, como un molde.

Luego ese molde se utiliza para crear la cantidad que necesitemos de los `objetos` de esa `clase`.


```python
# Molde
class Perro:
    # Atributos de clase
    cabeza = 1
    patas = 4

# Creando un objeto - Instanciar un objeto
firulais = Perro()
choco = Perro()
puppy = Perro()

print(firulais.cabeza)
print(choco.cabeza)
print(puppy.patas)
```

## Definiendo atributos
Es importante distinguir que existen dos tipos de atributos:

`Atributos de clase`: Son atributos cuyos valores definen a la clase, por lo tanto ser√°n comunes para todos los objetos. Son √∫tiles para cosas que todos los objetos comparten, como especie = "canino".

`Atributos de instancia`: Son atributos cuyos valores son espec√≠ficos de los objetos creados (la instancia). En nuestro caso, de ambas mascotas tiene el atributo `nombre`, pero cada una tiene un valor diferente, que se define al crear el objeto. (Mostrar los de instancia)


```python
# Molde
class Perro:
    # Atributos de clase
    especie = 'canina'
    
    # Metodo constructor para definir Atributos de instancia
    def __init__(self, nombre, raza, color):
        self.nombre = nombre
        self.raza = raza
        self.color = color

# Crear  el objeto
perro1 = Perro('Firulais', 'mestizo', 'marron')
print(perro1.especie)
print(perro1.raza)
print(perro1.color)

#  Crear otro objeto:
perro2 = Perro('Rompe', 'delmer', 'negro')
print(perro2.nombre)
print(perro2.especie)
```

## Definiendo m√©todos
En realidad cuando usamos `__init__` anteriormente ya est√°bamos definiendo un m√©todo, solo que uno especial. Uno que se ejecuta al momento de crear un objeto. A continuaci√≥n vamos a ver como definir m√©todos que le den alguna funcionalidad interesante a nuestra clase

```python
class Superheroe:
    # Atributo de clase
    especie = 'superhumanos'

    # Definir metodo constructor para definir sus atributos de instancia
    def __init__(self, nombre):
        self.nombre = nombre
    
    # Metodo
    def rayos_laser(self):
        print("tiro rayos laser rojos")
    
    def volar(self, km):
        print(f"Estoy volando a {km} km por hora")

    def alzar_objetos(self, objeto, kg):
        print(f"Puedo alzar un {objeto} de {kg} kg")


# Instanciar objetos
superheroe1 = Superheroe('Batman')
superheroe1.volar(120)

```

```python
class Coche:
    tipo = 'vehiculo'

    #Atributos de instancia
    def __init__(self, modelo, posicion = 0):
        self.modelo = modelo
        self.posicion = posicion

    # Definir otras acciones de la clase
    def acelerar(self, metros):
        self.posicion += metros
        print(f"El vehiculo modelo {self.modelo} acelera y avanza {metros} m. Nueva posicion: {self.posicion} m")
    
    def girar(self, direccion):
        print(f"El {self.modelo} avanza y gira a la direccion {direccion}")


coche1 = Coche('Toyota')
coche1.acelerar(1)
coche1.girar('izquierda')

    
```

### ü§ì Regla pr√°ctica

Si un atributo es parte esencial del objeto (modelo, posici√≥n, color, etc.) = atributo de clase

Si es algo opcional o derivado de una acci√≥n (√∫ltima direcci√≥n, velocidad temporal), puedes crearlo dentro de un m√©todo.

## Herencia
Ojo! Esto no es una herencia como tal, sino una relaci√≥n entre clases.
Lo que representa la herencia es que una clase hereda los atributos y m√©todos de otra clase. 

Dado que una clase hija hereda los atributos y m√©todos de la clase padre, nos puede ser muy √∫til cuando tengamos clases que se parecen entre s√≠ pero tienen ciertas particularidades

```python
class AnimalDeGranja:
    # atributos de clase:
    tipo = 'animal de granja'

    # Atributos de instancia
    def __init__(self, nombre, color, sonido):
        self.nombre = nombre
        self.color = color
        self.sonido = sonido

    # Metodos (Acciones de la clase)
    def hablar(self):
        print(f"Hablo y digo {self.sonido}")

    def presentarse(self):
        print(f"Hola, soy {self.nombre}")
    

class Gallina(AnimalDeGranja):
    huevos_por_dia = 1

    def poner_huevo(self):
        print(f"Estoy poniendo {self.huevos_por_dia} huevos por dia")


class Vaca(Gallina, Perro):
    patas = 4

    def mujir(self):
        print("Muuu")

gallina1 = Gallina('Hortencia', 'marron', 'corococo')
gallina1.hablar()
gallina1.presentarse()

vaca = Vaca('Lola', 'blanco y negro', 'Muuuu')
vaca.hablar()
vaca.mujir()
vaca.patas
```
