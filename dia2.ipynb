### `Operadores de comparacion`
- `>` - mayor que otro 
- `<` - menor que otro
- `>=` - mayor o igual que otro   
- `<=` - menor o igual que otro
- `==` - igual a otro     
- `!=` - desigual o distinto a otro

```python
numero1 = 19
numero2 = 20

nombre1 = 'Ana'
nombre2= 'ana'

# Validar una condicion
print(numero1 > numero2) # False
print(nombre1 == nombre2) # False
```

```python
#Usamos condicionales + operadores de comparacion
edad= 17

# Primera condicion
if edad < 18:
    print("Es menor de edad")
# segunda condicion
else:
    print("Es mayor de edad")
```

### `Operadores l칩gicos`
- `and` (y) - ambas condiciones deben ser verdaderas    
- `or` (o) - al menos una de las condiciones debe ser verdadera
- `not` (no) - invierte el valor de la condicion

```python
edad_laboral = 64
trabaja = True

# primera condicion
if edad_laboral < 18:
    print('Estudiante, menor de edad')

# segunda condicion
elif edad_laboral <= 65 and not trabaja:
    print('activo laboralmente')

# tercera condicion
elif edad_laboral < 65 and trabaja == False:
    print('Estudiante')
    
# cuarta y ultima condicion
else:
    print('jubilado')
```

```python
def verificar_edad_laboral(edad_laboral, trabaja):
    # primera condicion
    if edad_laboral < 18:
        print('Estudiante, menor de edad')
    # segunda condicion
    elif edad_laboral <= 65 or trabaja == True:
        print('activo laboralmente')
    # tercera condicion
    elif edad_laboral < 65 and trabaja == False:
        print('Estudiante')       
    # cuarta y ultima condicion
    else:
        print('jubilado')

verificar_edad_laboral(edad_laboral, trabaja)
```

### Definici칩n 

While es una palabra reservada que nos permite ejecutar una serie de instrucciones mientras una condicion se cumpla. Al momento que la condicion deje de ser cierta el bucle va parar.

```python
numero_a_adivinar= 20
numero_ingresado = 10

# While + Condicion
while numero_ingresado != numero_a_adivinar:
    numero_ingresado = int(input("Ingresa un numero"))
    print(numero_ingresado)
```

### Bucles infinitos

Un bucle infinito sucede cuando la condicion siempre es cierta y por lo tanto el bucle no termina de ejecutarse. "Break" es una palabra reservada que nos permite salir de un bucle.

```python
numero = 30

while True: 
    numero_a_ingresar = int(input("Ingresa un numero"))
    if numero_a_ingresar == numero:
        print("El nro es correcto")
        break
    else:
        print("El nro es incorrecto, intentalo de vuelta")
```
